```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px

matches = pd.read_csv("./superlig.csv", index_col=0)
```

```python
#@title En İyi Atak Takımları
overall_team = matches.sort_values('Overall', ascending=False)
plt.figure(figsize=(6,4),dpi=200)
sns.barplot(data=overall_team,y='Name',x='Attack')
plt.title("En İyi Atak Takımları")
```

```python
#@title En İyi Orta Saha Takımları
overall_team = matches.groupby('Overall').first().sort_values('Overall',ascending=True)[0:10]
plt.figure(figsize=(6,4),dpi=200)
sns.barplot(data=overall_team,y='Name',x='Midfield')
plt.title("En İyi Orta Saha Takımları")
```

```python
#@title En İyi Defans Takımları
overall_team = matches.groupby('Overall').first().sort_values('Overall',ascending=True)[0:10]
plt.figure(figsize=(6,4),dpi=200)
sns.barplot(data=overall_team,y='Name',x='Defence')
plt.title("En İyi Defans Takımları")
```

```python
#@title En Prestiji Takımlar
overall_team = matches.groupby('Overall').first().sort_values('Overall',ascending=True)[0:10]
plt.figure(figsize=(6,4),dpi=200)
sns.barplot(data=overall_team,y='Name', x='DomesticPrestige')
plt.title("En Prestiji Takımlar")
```
